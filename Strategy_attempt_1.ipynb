```python
df1 = pd.read_csv('/content/drive/MyDrive/data_4yr.csv')
df1.drop(columns='Unnamed: 0', inplace=True,axis=1)
```

```python
# logic behind this code snippet is we are considering only a 5% probably move in index after taking the trade.
# so the remaining strike prices are useless and are filtered out.
# Then we will try to use this data to get relevant ratios and stuff

def get_unique(numbers):
    unique = []
    for number in numbers:
        if number not in unique:
            unique.append(number)
    return unique

date_lis = list(df1['Date'])
dates = get_unique(date_lis)

df3 = pd.DataFrame()
for i in dates:
  df2 = pd.DataFrame()
  df2 = df1[df1['Date'] == i]
  p = df2['Underlying Value']*0.95

  df2.drop(df2[(df2['Strike Price'] < p)].index, inplace=True)
  df3 = pd.concat([df3,df2])

df4 = pd.DataFrame()
for i in dates:
  df2 = pd.DataFrame()
  df2 = df3[df3['Date'] == i]
  q = df2['Underlying Value']*1.05
  df2.drop(df2[(df2['Strike Price'] > q)].index, inplace=True)
  df4 = pd.concat([df4,df2])

df4.to_csv('updated_1.csv') 
df4.reset_index(inplace=True)
df4.drop(columns=['index'], inplace=True,axis=1)
df4
```
